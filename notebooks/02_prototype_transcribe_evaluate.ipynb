```python
import os
import openai
import speech_recognition as sr

openai.api_key = "xxxxxxxxxx"

# Archivo de audio
archivo = r"D:\Amatech\record_out.wav"
recognizer = sr.Recognizer()

try:
    with sr.AudioFile(archivo) as source:
        recognizer.adjust_for_ambient_noise(source, duration=0.5)
        print("Escuchando el audio...")
        audio = recognizer.record(source)

    text = recognizer.recognize_google(audio, language="es-MX").lower()
    print(f"\nTexto reconocido en la llamada:\n{text}")

except sr.UnknownValueError:
    print("No se pudo reconocer ningun texto en el audio.")
    text = ""
except sr.RequestError as e:
    print(f"Error al conectar con el servicio de reconocimiento: {e}")
    text = ""
except FileNotFoundError:
    print(f"No se encontró el archivo: {archivo}")
    text = ""

def analyze_call(call_text):
    pre_prompt = f"""
Evalúa la siguiente transcripción según esta rúbrica:
1. Saludo y presentación
2. Verificación del cliente
3. Escucha activa
4. Identificación de la necesidad
5. Conocimiento del producto/servicio
6. Ofrecimiento de solución o alternativa
7. Manejo de objeciones
8. Empatía y tono
9. Cierre y despedida
10. Cumplimiento del protocolo

Devuelve el resultado en formato JSON con las claves:
criterio, puntaje (1–5), comentario.

Transcripción:
{call_text}
"""

    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Eres un analista experto en evaluación de calidad de atención telefónica."},
            {"role": "user", "content": pre_prompt}
        ],
        temperature=0.4
    )

    return response.choices[0].message.content

if text.strip():
    analisis = analyze_call(text)
    print("\n--- RESULTADO DEL ANALISIS ---\n")
    print(analisis)
else:
    print("No se pudo realizar el análisis: no se reconocio texto válido.")
```
